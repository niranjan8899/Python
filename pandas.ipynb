```python
data=pd.read_csv("/Users/niranjansmac/Downloads/train_and_test2.csv")
data.head()
```

```python
#cleaning the data without copying the data into another variable
data.dropna(inplace=True,axis=0)
data.isna().sum()
```

```python
#2.replacing missing values
data=pd.read_csv("/Users/niranjansmac/Downloads/train_and_test2.csv",usecols=['sibsp','Passengerid','Fare','2urvived'])

#replacing NaN values using mean
mean_values=data['Fare'].mean()
print(mean_values)
data2=data.fillna(mean_values)
data2.head(21)
```

```python
#replacing NAN values with median
data=pd.read_csv("/Users/niranjansmac/Downloads/train_and_test2.csv")
median_values=data['Fare'].median()
print(median_values)

data2=data.fillna(median_values)
data2
```

```python
#3.Duplicate data
data=pd.read_csv("/Users/niranjansmac/Downloads/train_and_test2.csv",usecols=['sibsp','Passengerid','Fare','2urvived'])
print(data.shape)
data2=data.drop_duplicates()
print(data2.head(11))
data2.shape
```

```python
#renaming columns
data=pd.read_csv("/Users/niranjansmac/Downloads/train_and_test2.csv",usecols=['sibsp','Passengerid','Fare','2urvived'])
data.rename(columns={'Passengerid':'id'},inplace=True)
data.head()
```
