```python
import matplotlib.pyplot as plt
```

```python
#types of matplotlib
```

```python
#1.matplotlib line plot
x=[41,52,62,72,34,67]
y=[23,1,22,44,99,100]
plt.bar(x,y)
plt.show()
```

```python
#2.histogram
x=[41,52,62,72,34,67,41,52,62,72,34,67,67,41,52,62,62,62,62,62,62,72,34,67,67,41,72,34,67,67,41,72,34,67,67,41]
plt.hist(x)
plt.show()
```

```python
#3.scatter plot
x=[41,52,62,72,34,67,23,1,22,44,99,100]
y=[23,1,22,44,99,34,67,23,1,22,88,100]
plt.scatter(x,y)
plt.show()
```

```python
#4.pie chart
subjects=['Telugu','English','Maths','Science']
marks=[77,23,81,100]
plt.pie(marks,labels=subjects,autopct='%1.1f%%',startangle=90)
plt.show()
```

```python
#5.area plot
x=[41,52,62,72,34,67]
y,y1=[23,1,22,44,99,34],[41,52,62,72,34,67]
plt.fill_between(x,y,y1, color='red',alpha=0.4,label='Area 1-2')
plt.plot(x,y,label='Line 1',marker='o')
plt.plot(x,y1,label='Line 1',marker='o')
plt.xlabel('x-axis'),plt.ylabel('y-axis'),plt.title('Area chart')
plt.legend(), plt.show()
```

```python
#introduction to figures
fig=plt.figure(figsize=(4,4),facecolor='lightblue',layout='constrained')
fig.suptitle('figure')
ax=fig.add_subplot()
ax.set_title('Axes',loc='left',fontstyle='oblique',fontsize='medium')
```

```python
fig,a=plt.subplots(2,2)
import numpy as np
x=np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title("square")
a[0][1].plot(x,x+x)
a[0][1].set_title('add')
a[1][0].plot(x,x-x)
a[1][0].set_title('subtract')
a[1][1].plot(x,x**x)
a[1][1].set_title('pow')
plt.show()
```
